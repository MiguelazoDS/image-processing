### Introducción: 

Primero se hará una prueba para comparar los diferentes módulos de Python usados para el procesamiento de imágenes, para elegir uno que se usará a lo largo de todo el práctico.

En el siguiente código se intentará abrir y graficar mediante **matplotlib** una misma imagen con **PIL**, **OpenCV** y **SimpleITK**.

```python
import numpy as np
import cv2 as cv
from PIL import Image
import SimpleITK as sitk
from matplotlib import pyplot as plt

%matplotlib inline

#Leemos la imagen usando 3 módulos diferentes
image_PIL=Image.open("fcefyn.jpg")
image_SITK=sitk.ReadImage("fcefyn.jpg")
image_CV=cv.imread("fcefyn.jpg")

print("Vemos el tipo de cada variable\n")
print(("Tipo PIL: %s\n,Tipo SITK: %s\n,Tipo CV: %s") % (type(image_PIL), type(image_SITK), type(image_CV)))

#Ponemos 3 gráficos uno al lado del otro con su propio título.
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=True)
fig.set_figheight(10)
fig.set_figwidth(15)
axs[0].imshow(image_CV)
axs[0].set_title("Imagen con OpenCV")
axs[2].imshow(image_SITK)
axs[2].set_title("Imagen con SimpleITK")
axs[1].imshow(image_PIL)
axs[1].set_title("Imagen con PIL")
plt.show()
```

Vemos que el tipo que se obtiene con SimpleITK no nos permite graficar y en OpenCV hay un problema con los colores. En el siguiente código tratamos de convertir el tipo de la variable anterior e intentamos graficar nuevamente. 

```python
image_SITK_array=np.asarray(image_SITK)
print(type(image_SITK_array))
print(image_SITK_array)
plt.imshow(image_SITK_array)
plt.show()
```

Aunque el tipo ahora es igual que el obtenido directamente con OpenCV no podemos todavía graficar. Usamos entonces la función propia de SimpleITK llamada "GetArrayFromImage" para la conversión de tipo.

```python
image_SITK_array=sitk.GetArrayFromImage(image_SITK)
print(type(image_SITK_array))
print(image_SITK_array)
plt.imshow(image_SITK_array)
plt.title("Imagen con SimpleITK")
plt.show()
```

De esta forma se pudo graficar correctamente. El problema con "asarray" de numpy es que la matriz queda incompleta.

El inconveniente con los colores en OpenCV, según información obtenida de [How to Display a Matplotlib RGB Image](https://www.pyimagesearch.com/2014/11/03/display-matplotlib-rgb-image/), es que cuando OpenCV representa las imagenes RGB en un arreglo multidimencional, lo hace de manera invertida, es decir que lo que en realidad está representando es una imagen BGR. Para invertir la imagen se hace como se indica en el siguiente código.

```python
plt.imshow(cv.cvtColor(image_CV, cv.COLOR_BGR2RGB)) 
plt.title("Imagen con OpenCV corregida")
plt.show()
```

Para el resto del práctico se eligirá OpenCV ya que no necesita conversión a array y la representación en BGR puede resolverse muy fácilmente.

## Ejercicio 1:

### Realice y grafique una imagen con una cruz negra y fondo blanco.

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2 as cv

#Creamos una matriz de 128x128 con unos.
matriz=np.ones((128,128))

#Multiplicamos la matriz por 255 para que nos de el color blanco
matriz=255*matriz

#Con 44:84 estamos indicando la cantidad de filas, de 44 hasta 84 en intervalo abierto,
#las multiplicamos por cero (para obtener el negro) y guardamos el resultado en el mismo lugar.
matriz[44:84]=0*matriz[44:84]

#Del mismo modo con 44:84 obtenemos la misma cantidad de columnas.
matriz[:,44:84]=0*matriz[:,44:84]

#Graficamos
plt.imshow(matriz, cmap='gray', vmin=0 ,vmax=255)
plt.show()
```

## Ejercicio 2:

### Realice y grafique una imagen con una cruz blanca y fondo negro (negativo de la primera).

```python
#Recorremos cada uno de los valores de la matriz y cambiamos los 255 por 0 y viceversa.
for i in range(0, len(matriz[0])):
    for j in range(0, len(matriz[:,0])):
        if matriz[i][j]==0:
            matriz[i][j]=255
        else:
            matriz[i][j]=0

#Graficamos
plt.imshow(matriz, cmap='gray', vmin=0 ,vmax=255)
plt.show()
```

## Ejercicio 3:

### Realice y grafique, una imagen con cuatro bandas horizontales de intensidad homogénea decreciente de 1 a 0 de arriba abajo.

```python
#Creamos una matriz de 128x128 con unos.
matriz=np.ones((128,128))

#Como se realizó anteriormente y modificando los rangos convenientemente tenemos. Ahora el blanco
#está representado por 1 y el negro por 0, los valores entre estos son diferentes intensidades de grises
matriz[32:64]=0.66*matriz[32:64]
matriz[64:96]=0.33*matriz[64:96]
matriz[96:128]=0*matriz[96:128]

#Modificamos el valor máximo a 1 y graficamos
plt.imshow(matriz, cmap='gray', vmin=0 ,vmax=1)
plt.show()
```

## Ejercicio 4:

### Construir y graficar una imagen 128x128 dividida en cuatro cuadrados de colores: rojo, azul, amarillo y uno a elección.

```python
#Creamos una matriz de 128x128x3 con unos.
matriz=np.ones((128,128,3))

#En las 3 siguientes líneas tomamos 64 filas (del 0 al 63) y 64 columnas (del 0 al 63).
#Para cada valor matriz[fila,columna] existe un array de 3 elementos, que se acceden mediante
#0, 1 y 2 que representan los valores de color rojo, verde y azul respectivamente.
#Al multiplicar por 0 el valor en las últimas dos líneas estamos "apagando" los colores verde
#y azul, dejando solo el rojo. El valor que multiplica la matriz en la línea uno, va desde 
#0 a 1 regulando su intensidad
matriz[0:64,0:64,0]=0.8*matriz[0:64,0:64,0]
matriz[0:64,0:64,1]=0*matriz[0:64,0:64,1]
matriz[0:64,0:64,2]=0*matriz[0:64,0:64,2]

#De manera similar tomamos otra sección de la imagen y dejamos "encendido" solo el color azul
#con una intensidad de 0.6
matriz[0:64,64:128,0]=0*matriz[0:64,64:128,0]
matriz[0:64,64:128,1]=0*matriz[0:64,64:128,1]
matriz[0:64,64:128,2]=0.6*matriz[0:64,64:128,2]

#Para formar el amarillo tenemos que combinar los colores rojo y verde.
matriz[64:128,0:64,0]=0.9*matriz[64:128,0:64,0]
matriz[64:128,0:64,1]=0.9*matriz[64:128,0:64,1]
matriz[64:128,0:64,2]=0*matriz[64:128,0:64,2]

#Con alguna herramienta online que nos permite fácilmente elegir un color y ver sus valores RGB
#podemos mostrar cualquier color. Se eligió mostrar un color celeste.
matriz[64:128,64:128,0]=0.4*matriz[64:128,64:128,0]
matriz[64:128,64:128,1]=0.7*matriz[64:128,64:128,1]
matriz[64:128,64:128,2]=0.9*matriz[64:128,64:128,2]

#Graficamos
plt.imshow(matriz, cmap = 'gray',vmin=0 ,vmax=1)
plt.show()
```

## Ejercicio 5:

### Construir y graficar una imagen a color a su gusto.

```python
#Creamos una matriz de 128x128x3 con unos.
matriz=np.ones((128,128,3))

#Se dibuja un cuadrado donde va variando la intensidad desde un valor muy pequeño empezando
#desde arriba, a un valor igual a 0.6, 0.8 y 0.9
k=1
for i in range(0,80):
    for j in range(0,80):
            factor=k/6400
            matriz[24+i][24+j][0]=0.6*factor*matriz[24+i][24+j][0]
            matriz[24+i][24+j][1]=0.8*factor*matriz[24+i][24+j][1]
            matriz[24+i][24+j][2]=0.9*factor*matriz[24+i][24+j][2]
            k+=1

#Graficamos
plt.imshow(matriz, cmap = 'gray',vmin=0 ,vmax=1)
plt.show()
```

## Ejercicio 6:

### Abrir una imagen guardada en un archivo, convertirla en Imagen monocroma, y graficar ambas imágenes a la par, colocarle título a ambas.

```python
#Abrimos la imagen
imagen=cv.imread('Córdoba.jpg')

#Creamos los sub-gráficos.
fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True)
fig.set_figheight(10)
fig.set_figwidth(15)

#Graficamos la imagen original recordando convertir la imagen de BGR a RGB.
axs[0].imshow(cv.cvtColor(imagen, cv.COLOR_BGR2RGB), cmap='gray')
axs[0].set_title("Imagen original")

#Anulamos las componentes roja y verde de la imagen
azul=imagen[:,:,2]

#Graficamos usando una paleta de azules con el agregado de _r 
#para que no se muestren invertidos los colores.
axs[1].imshow(azul, cmap='Blues_r')
axs[1].set_title("Imagen monocroma")
plt.show()

print("Podemos ver que la matriz que corresponde a la imagen monocromática solo es de dos dimensiones")
print("azul.shape: ",azul.shape)
```

```python
#Seleccionamos las filas y columnas donde se encuentra la catedral de manera aproximada viendo los ejes.
recorte=azul[50:320,110:420]
plt.imshow(recorte,cmap='Blues_r')
plt.show()
```

## Ejercicio 8:

### Graficar la imagen monocroma del ejercicio 6 corrompida por un cuadrado blanco y guardarla en un archivo con formato de imagen.

```python
#Seleccionamos un cuadrado mediante ciclos for y a cada punto lo hacemos igual a 255 (blanco).
for i in range(100,300):
    for j in range(100,300):
        azul[i][j]=255
        
#Guardamos la imagen poniéndole un nombre diferente para evitar que se sobreescriba.
cv.imwrite('imagen_corrompida.jpg',azul)

#Graficamos
plt.imshow(azul,cmap='Blues_r')
plt.show()
```
