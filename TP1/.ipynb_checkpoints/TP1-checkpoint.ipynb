Primero se hará una prueba para comparar los diferentes módulos de Python usados para el procesamiento de imágenes, para elegir uno que se usará a lo largo de todo el práctico.

En el siguiente código se intentará abrir y graficar mediante **matplotlib** una misma imagen con **PIL**, **OpenCV** y **SimpleITK**.

```python
import numpy as np
import cv2 as cv
from PIL import Image
import SimpleITK as sitk
from matplotlib import pyplot as plt

%matplotlib inline

image_PIL=Image.open("fcefyn.jpg")
image_SITK=sitk.ReadImage("fcefyn.jpg")
image_CV=cv.imread("fcefyn.jpg")

print("Vemos el tipo de cada variable abriendo la imagen con diferentes módulos\n")
print(("Tipo PIL: %s\n,Tipo SITK: %s\n,Tipo CV: %s") % (type(image_PIL), type(image_SITK), type(image_CV)))

#Ponemos 3 gráficas una al lado de la otra con su propio título.
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=True)
fig.set_figheight(10)
fig.set_figwidth(15)
axs[0].imshow(image_CV)
axs[0].set_title("Imagen con OpenCV")
axs[2].imshow(image_SITK)
axs[2].set_title("Imagen con SimpleITK")
axs[1].imshow(image_PIL)
axs[1].set_title("Imagen con PIL")
plt.show()

#print(image_PIL, image_CV)
```

Vemos que el tipo que se obtiene con SimpleITK no nos permite graficar y en OpenCV hay un problema con los colores. En el siguiente código tratamos de convertir el tipo de la variable anterior e intentamos graficar nuevamente. 

```python
image_SITK_array=np.asarray(image_SITK)
print(type(image_SITK_array))
print(image_SITK_array)
plt.imshow(image_SITK_array)
plt.show()
```

Aunque el tipo ahora es igual que el obtenido directamente con OpenCV no podemos todavía graficar. 
Usamos entonces la función propia de SimpleITK llamada "GetArrayFromImage" para la conversión de tipo.

```python
image_SITK_array=sitk.GetArrayFromImage(image_SITK)
print(type(image_SITK_array))
print(image_SITK_array)
plt.imshow(image_SITK_array)
plt.title("Imagen con SimpleITK")
plt.show()
```

De esta forma se pudo graficar correctamente. El problema con "asarray" de numpy es que la matriz queda incompleta.

El problema con OpenCV según información obtenida de 
[How to Display a Matplotlib RGB Image](https://www.pyimagesearch.com/2014/11/03/display-matplotlib-rgb-image/)
es que cuando OpenCV representa las imagenes RGB en un arreglo multidimencional, lo hace de manera invertida, es 
que lo que en realidad está representando es una imagen BGR. Para invertir la imagen se hace como se indica en el
siguiente código.

```python
plt.imshow(cv.cvtColor(image_CV, cv.COLOR_BGR2RGB)) 
plt.title("Imagen con OpenCV corregida")
plt.show()
```

Para el resto del práctico se eligirá OpenCV ya que no necesita conversión a array y la representación en BGR puede resolverse muy fácilmente.

## Ejercicio 1:

### Realice y grafique una imagen con una cruz negra y fondo blanco.

```python
from matplotlib import pyplot as plt
import cv2 as cv
#Creamos una matriz de 128x128 con unos.
matriz=np.ones((128,128))
#Multiplicamos la matriz por 255 para que nos de el color blanco
matriz=255*matriz
#Con 44:84 estamos indicando la cantidad de filas, las multiplicamos por cero 
#(para obtener el negro) y guardamos el resultado en el mismo lugar.
matriz[44:84]=0*matriz[44:84]
#Del mismo modo con 44:84 obtenemos la cantidad de columnas.
matriz[:,44:84]=0*matriz[:,44:84]

plt.imshow(matriz, cmap='gray', vmin=0 ,vmax=255)
plt.show()
```

## Ejercicio 2:

### Realice y grafique una imagen con una cruz blanca y fondo negro (negativo de la primera).

```python
#Recorremos cada uno de los valores de la matriz y cambiamos los 255 por 0 y viceversa.
for i in range(0, len(matriz[0])):
    for j in range(0, len(matriz[:,0])):
        if matriz[i][j]==0:
            matriz[i][j]=255
        else:
            matriz[i][j]=0
               
plt.imshow(matriz, cmap='gray', vmin=0 ,vmax=255)
plt.show()
```
